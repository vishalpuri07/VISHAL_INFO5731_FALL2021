```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("interaction_analysis.csv")
```

```python
df = df1[['text', 'user1','user2','user3','user4','user5']]
df = df[df['text'].notna()]
df = df.drop_duplicates(subset = ["text"])
df = df[df['user1'].notna()]
df
```

```python
df_username = pd.DataFrame()
```

```python
df_username["username"] = df["user1"].append(df["user2"])
df_username
```

```python
df_username.value_counts().nlargest(n=15).sort_values().drop("@").plot(kind = 'barh');
plt.xlabel("Count", labelpad = 15, fontsize = 15);
plt.ylabel("Username",labelpad = 15 ,fontsize = 15);
plt.title("Most popular username", fontsize = 20);
```
