```python
##### CiteSeerX - Smart Health Query #####

# Import necessary Libraries. to create beautifulsoup and dataframe.
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd

# Create Empty List so that the data is stored on this list later.
Abstracts = []

for page in range(0,500,10): # each page has 10 results. 
#THE WEBPAGE PROVIDES ONLY TOP 500 LIST TO THE PUBLIC.
    link =  urllib.request.urlopen(f"https://citeseerx.ist.psu.edu/search?q=smart+health&t=doc&sort=rlv&start={page}")


# The URL Above is searching for term "Smart Health" in the citeseerx website.
    soup = BeautifulSoup(link) # Now the soup is ready which is a parsed HTML type document. 
    for i in soup.findAll(class_ = "snippet"): # class for finding the Abstract
      Abstracts.append(i.get_text(strip = True)) # appends for all pages of the soup
      page = page + 1 # Move to next page. 
    
    
# To clean the white spaces and new line character use strp()    
for i in range(len(Abstracts)): 
    Abstracts[i] = str(Abstracts[i].replace("...","").strip()) # removing leading by and whitespaces.
```

```python
# Convert list to dataframe to convert it to csv file. 
abstracts_df = pd.DataFrame(Abstracts, columns = ["Abstracts"]) # Converting to dataframe
abstracts_csv = abstracts_df.to_csv(index = False) # Converting to .csv file
abstracts_csv # Quick sanity check.
```

```python
##### PART 1 - Basic feature extraction using text data##### 

# Link to the website after clickig on "Raw" in order to get cleaner data.

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords'); #download stopwords if necessary


link =  urllib.request.urlopen("https://raw.githubusercontent.com/unt-iialab/info5731-fall2021/main/assignments/01-05-1%20%20Adams%20v%20Tanner.txt")
soup = BeautifulSoup(link) # Now the soup is ready which is a parsed HTML type document.

# output the text to a text file type object to perform same operations as a text file.
file_contents = (soup.get_text().replace("\n", "")).replace("\r", "").strip()

# For total sentences just split the file with period.
print("Total Sentences:        ", len(file_contents.split(".")))

# For total words in the file just split the file with space after replacing the periods.
print("Total words:            ", len(file_contents.replace("."," ").split()))

# To get the total characters simply print the length of the texts.
print("Total Characters        ",len(file_contents))

# To get the total stops count the period in the file.
print("Total stops:            ", file_contents.count("."))

# Average word length = sum of length of words / length of total sentences.
average_word_length = sum(len(word) for word in file_contents.split()) / len(file_contents.split())
print("Average word length     ", round(average_word_length))

# For Stopwords use the stopwords provided by nltk. 
stop_words_list = []
stop_words = stopwords.words('english') # get all the stopwords in english language.
for word in file_contents.split():
  if word in stop_words:
    stop_words_list.append(word)
print("Total Stopwords         ", len(stop_words_list))


# For number of special characters:
special_char = '[ @ + - _ ! # = $ % ; ^ & * ( | ) < > ? / \ | } { ~ : ]'
count_special = []
for char in file_contents:
  if char in special_char:
    count_special.append(char)
print("Total special character ", len(count_special))


# For number of numerics. 
numerics_count = 0
for char in file_contents:
    if char.isdigit():
        numerics_count += 1
print("Total numerics          ",numerics_count)


# For total number of upercase words.
uppercase_count = 0
for char in file_contents:
    if char.isupper():
        uppercase_count += 1
print("Total Upercase Words    ",uppercase_count)
```

```python
##### PART 2 - Basic Text Pre-processing of text data #####

# Same code below to get the text from the webpage. 

link =  urllib.request.urlopen("https://raw.githubusercontent.com/unt-iialab/info5731-fall2021/main/assignments/01-05-1%20%20Adams%20v%20Tanner.txt")
soup = BeautifulSoup(link) # Now the soup is ready which is a parsed HTML type document.

# output the text to a text file type object to perform same operations as a text file.
file_contents = (soup.get_text().replace("\n", "")).replace("\r", "").strip() # remove new line and tab.


# 1. Lowercase
file_contents = file_contents.lower() # Note that if you copy the code above for uppercase down after this line, it will result in 0 total upper case characters. 

# 2. Punctuation removal
file_contents = file_contents.replace('[^\w\s]', "") # From professor notes.

# 3. Stopwords Removal
file_contents = file_contents = " ".join(word for word in file_contents.split() if word not in stop_words)
# For quick sanity check run the code above for stopwords and corretcly so it gives 0 as output for total stopwords.


# 4. Frequent words removal
# Use dictionary to finds most repeating words and its counts.
import collections
common_dict = collections.Counter(file_contents.split())
most_common_keys = {key for key, val in common_dict.items() if val > 15}  # If anything repeats over 15 times remove it.

# Remove all the words that are in the common_keys items above.
file_contents = " ".join(word for word in file_contents.split() if word not in most_common_keys)


# 5. Rare words removal

common_dict = collections.Counter(file_contents.split())
least_common_keys = [key for key, val in common_dict.items() if val == 1]
# Upon looking more into there are many values that are only repeated once
least_common_keys.sort(reverse = False)
# However there are digits with special characters attached such as *741, 1812a and sortings them brings these ahead in the list, so lets remove those. These are top 100 words or so.
least_common_keys_list = least_common_keys[100:]

file_contents = " ".join(word for word in file_contents.split() if word not in least_common_keys)

# 6. Spelling Correction

# for this import TextBlob
from textblob import TextBlob
for words in file_contents.split():
  words = str(TextBlob(words).correct())

# 7. Tokenization.
import spacy
nlp = spacy.load('en_core_web_sm') #Load English tokenizer and tagger.
for entity in nlp(file_contents).ents:
    print(entity.text + ' - ' + entity.label_ + ' - ' + str(spacy.explain(entity.label_)))
```

```python
# 8. Stemming
import nltk
nltk.download('wordnet')

from nltk.stem import PorterStemmer
st = PorterStemmer()
file_contents = " ".join([st.stem(word) for word in file_contents.split()])

# 9. Lemmatization. 
from textblob import Word
file_contents = " ".join([Word(word).lemmatize() for word in file_contents.split()])
file_contents
```

```python
##### 1. POS Tagging #####
""" spaCy is a python libraries that works particularly well with POS Tagging.
 Import spacy. If not already installed use "pip install spacy" to download spaCy
"""
# pip install spacy ----> install spacy if required
import spacy
nlp = spacy.load('en_core_web_sm') #Load English tokenizer and tagger.

# Create Empty list so that all the nouns , adjectives, verb and adverb can be stored.
nouns = [] # NOUN
adjectives = [] # ADJ
verbs = [] # VERB
adverbs = [] # ADV

for content in file_contents.split(): # Loop each content from the file_contents and find the tags and append them 
  text = nlp(content)
  for token in text:
    if token.pos_ == 'NOUN': # For nouns use tag: NOUN
      nouns.append(token)
    if token.pos_ == 'ADJ': # For adjectives use tag: ADJ
      adjectives.append(token)
    if token.pos_ == "VERB": # For verbs use tag: VERB
      verbs.append(token)
    if token.pos_ == "ADV": # For adjectives use tag: ADJ
      adverbs.append(token)
# print the count for each. See contents by simply printing it such as print(nouns), print(verbs), etc.
print("Total Number of Nouns:       ", len(nouns))
print("Total Number of Verbs:       ", len(verbs))
print("Total Number of Adjectives:  ", len(adjectives))
print("Total Number of Adverbs:     ", len(adverbs))
```
